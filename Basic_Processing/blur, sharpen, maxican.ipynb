```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_path = './data/car.jpeg'

image = cv2.imread(image_path, 0)

plt.imshow(image, 'gray')
plt.title('original image')
plt.show()
```

```python
image_blurred = np.hstack([
    cv2.blur(image, (3, 3)),
    cv2.blur(image, (5, 5)),
    cv2.blur(image, (9, 9))
])

plt.imshow(image_blurred, 'gray')
plt.title('image_blurred')
plt.show()
```

```python
print(cv2.blur(image, (3, 3)))
print(cv2.blur(image, (3, 3)).shape)

print(image_blurred)
print(image_blurred.shape)
```

```python
image_gaussian = np.hstack([
    cv2.GaussianBlur(image, (3, 3), 0),
    cv2.GaussianBlur(image, (5, 5), 0),
    cv2.GaussianBlur(image, (9, 9), 0)
])

plt.imshow(image_gaussian, 'gray')
plt.title('image_gaussian')
plt.show()
```

```python
image1 = cv2.imread(image_path, 1)
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)

filter = np.array([[-1, -1, -1],
                   [-1, 9, -1],
                   [-1, -1, -1]])

image_sharp = cv2.filter2D(image1, -1, filter)

plt.imshow(image_sharp)
plt.title('image_sharp')
plt.show()
```

```python
filter = np.array([[0, 0, -1, 0, 0],
                   [0, -1, -2, -1, 0],
                   [-1, -2, 16, -2, -1],
                   [0, -1, -2, -1, 0],
                   [0, 0, -1, 0, 0]])

image_maxican = cv2.filter2D(image1, -1, filter)

plt.imshow(image_maxican)
plt.title('image_maxican')
plt.show()
```
