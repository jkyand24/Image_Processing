```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np

image_path = './data/car.jpeg'

image = cv2.imread(image_path, 1)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
filter = np.array([[3, -2, -3],
                   [-4, 8, -6],
                   [5, -1, 0]])

image_custom = cv2.filter2D(image, -1, filter)

plt.imshow(image_custom)
plt.title('image_custom')
plt.show()
```

```python
filter = np.array([[0.272, 0.534, 0.131],
                   [0.119, 0.686, 0.168],
                   [0.393, 0.769, 0.189]])

image_sepia = cv2.transform(image, filter)

plt.imshow(image_sepia)
plt.title('image_sepia')
plt.show()
```

```python
filter = np.array([[0, 1, 0],
                   [0, 0, 0],
                   [0, -1, 0]])

image_emboss = cv2.filter2D(image, -1, filter)

plt.imshow(image_emboss)
plt.title('image_emboss')
plt.show()

image_emboss += 128

plt.imshow(image_emboss)
plt.title('image_emboss')
plt.show()
```
