```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_path = './data/apple.jpeg'

image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

imageg = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
imageg_10x10 = cv2.resize(imageg, (10, 10))
print(imageg_10x10.flatten().shape)

plt.imshow(imageg_10x10, 'gray')
plt.title('imageg_10x10')
plt.show()
```

```python
image_10x10 = cv2.resize(image, (10, 10))
print(image_10x10.flatten().shape)

image_255x255 = cv2.resize(image, (255, 255))
print(image_255x255.flatten().shape)

plt.imshow(image_10x10)
plt.title('image_10x10')
plt.show()

plt.imshow(image_255x255)
plt.title('image_255x255')
plt.show()
```
