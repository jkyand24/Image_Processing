```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np

image_path = './data/apple.jpeg'

image = cv2.imread(image_path, 0)
```

Thresholding

```python
_, mask = cv2.threshold(image, 230, 255, cv2.THRESH_BINARY_INV)

image_list = np.hstack([image, mask])

plt.imshow(image_list, 'gray')
plt.title('image_list')
plt.show()
```

Dilation

```python
kernel = np.ones((3, 3), np.uint8)

image_dilation = cv2.dilate(mask, kernel)

image_list = np.hstack([image, mask, image_dilation])

plt.imshow(image_list, 'gray')
plt.title('image_list')
plt.show()
```

```python
image_dilation_10 = cv2.dilate(mask, kernel, iterations = 10)

image_list = np.hstack([image, mask, image_dilation_10])

plt.imshow(image_list, 'gray')
plt.title('image_list')
plt.show()
```

Erosion

```python
image_erosion = cv2.erode(mask, kernel)

image_list = np.hstack([image, mask, image_erosion])

plt.imshow(image_list, 'gray')
plt.title('image_list')
plt.show()
```

Opening, Closing, Gradient, Tophat

```python
image_opening = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
image_closing = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
gradient = cv2.morphologyEx(mask, cv2.MORPH_GRADIENT, kernel)
tophat = cv2.morphologyEx(mask, cv2.MORPH_TOPHAT, kernel)

image_list = np.hstack([image, mask, image_dilation, image_erosion, image_opening, image_closing, gradient, tophat])

plt.imshow(image_list, 'gray')
plt.title('image_list')
plt.show()
```
