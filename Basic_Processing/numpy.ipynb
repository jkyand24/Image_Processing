단일 객체 저장 및 불러오기

```python
import numpy as np
import os

array = np.arange(0, 10)
print(array)
```

```python
os.makedirs("./data", exist_ok = True)
np.save("./data/save.npy", array)
```

```python
result = np.load("./data/save.npy")
print(result)
```

복수 객체 저장 및 불러오기

```python
array1 = np.arange(0, 10)
array2 = np.arange(10, 20)

print(array1, array2)
```

```python
np.savez("save.npz", array1 = array1, array2 = array2)
result = np.load('save.npz')

data1 = result['array1']
data2 = result['array2']
print(result, data1, data2)
```

원소의 정렬

```python
array = np.array([5, 10, 20, 30, 2])
array.sort()
print(array)
```

```python
print(array[::-1])
```

```python
array = np.array([[5, 10, 20, 30, 2], [7, 3, 5, 62, 6]])
org_array = array.copy()

print(array)

array.sort(axis = 0)
print(array)

print(org_array)

org_array.sort(axis=1)
print(org_array)
```

linspace

```python
array = np.linspace(0, 10, 5)
print(array)
```

random.seed

```python
np.random.seed(4444)
print(np.random.randint(0, 10, (2, 3)))
```

array 복사

```python
array1 = np.arange(0, 10)
array2 = array1

array2[0] = 99

print(array1, array2)
```

```python
array3 = np.arange(0, 10)
array4 = array3.copy()

array4[0] = 99

print(array3, array4)
```

중복된 원소 제거

```python
array = np.array([1, 1, 2, 2, 3, 3, 4])
print(array)
temp = np.unique(array)
print(temp)
```

브로드캐스팅

```python
A = np.array([[1, 2, 3], [4, 5, 6]])
B = np.array([10, 20, 30])

C = A + B

print(C)
```

이미지 저장

```python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread("./data/cat.png")

plt.imshow(img)
plt.title('npy image')
plt.show()

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

np.save('./data/image.npy', img)

img_load = np.load('./data/image.npy')

plt.imshow(img_load)
plt.title('npy image loaded')
plt.show()
```
