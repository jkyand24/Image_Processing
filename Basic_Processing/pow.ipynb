```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('./data/image01.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    # image  : 255

plt.imshow(image)
plt.title('image')
plt.show()

gamma = 3
image_corrected = cv2.pow(image/255.0, gamma)
    # image_corrected  : 1.0
image_corrected = image_corrected * 255.0
    # image_corrected  : 255.0
"""
     image     255  ,
     1 , pow  ,     ,       
     255 ,   1  , pow .
       0~1 , 
     255 .
""" 

plt.imshow(image_corrected)
plt.title('image_corrected')
plt.show()
```
