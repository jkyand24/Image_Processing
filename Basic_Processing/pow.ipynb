```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('./data/image01.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    # image의 최대 픽셀값: 255

plt.imshow(image)
plt.title('image')
plt.show()

gamma = 3
image_corrected = cv2.pow(image/255.0, gamma)
    # image_corrected의 최대 픽셀값: 1.0
image_corrected = image_corrected * 255.0
    # image_corrected의 최대 픽셀값: 255.0
"""
    현재 image 행렬의 픽셀 값들이 최대 255일 수 있는데,
    픽셀값이 1 초과라면, pow의 지수에 따라, 운 나쁘면 픽셀값이 매우매우 커져, 이미지의 픽셀값 범위를 넘어갈 위험이 있으니 
    우선 255로 나눠주어, 가능한 최대값을 1로 만들어준 다음에야, pow를 적용한다.
    그 결과 픽셀값은 0~1의 범위이므로, 
    다시 255를 곱해준다.
""" 

plt.imshow(image_corrected)
plt.title('image_corrected')
plt.show()
```
