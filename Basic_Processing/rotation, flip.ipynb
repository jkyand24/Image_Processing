```python
import cv2
import matplotlib.pyplot as plt

image_path = './data/cat.png'
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
image_90 = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
image_180 = cv2.rotate(image, cv2.ROTATE_180)
image_270 = cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE)
```

```python
plt.imshow(image)
plt.title('image')
plt.show()

plt.imshow(image_90)
plt.title('image_90')
plt.show()

plt.imshow(image_180)
plt.title('image_180')
plt.show()

plt.imshow(image_270)
plt.title('image_270')
plt.show()
```

```python
image_lr = cv2.flip(image, 1)
image_ud = cv2.flip(image, 0)
```

```python
plt.imshow(image_lr)
plt.title('image_lr')
plt.show()

plt.imshow(image_ud)
plt.title('image_ud')
plt.show()
```
