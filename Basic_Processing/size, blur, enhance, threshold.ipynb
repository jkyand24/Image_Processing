```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image_path = './data/cat.png'

image = cv2.imread(image_path)

print(image)
```

```python
image_type = type(image)
image_height, image_width, image_channel = image.shape
print(image_type, image_height, image_width, image_channel)
```

```python
imagey = cv2.cvtColor(image, cv2.COLOR_BGR2YUV)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)


plt.imshow(image)
plt.title('original image')
plt.show()
```

```python
image_resize = cv2.resize(image, (224, 224))

image_height, image_width, image_channel = image_resize.shape
print("image_resize.shape:\n", image_height, image_width, image_channel)

plt.imshow(image_resize)
plt.title('resized image')
plt.show()
```

```python
cv2.imwrite('./data/resize_cat.png', image_resize)
```

```python
image_cropped = image[10:, :200]

plt.imshow(image_cropped)
plt.title('cropped image')
plt.show()
```

```python
image_blurred = cv2.blur(image, (45, 45))

plt.imshow(image_blurred)
plt.title('blurred image')
plt.show()
```

```python
kernel = np.ones((10, 10)) / 100

image_kernel = cv2.filter2D(image, -1, kernel)

plt.imshow(image_kernel)
plt.title('kerneled image')
plt.show()
```

```python
image_gaussianblur = cv2.GaussianBlur(image, (5, 5), 1)

plt.imshow(image_gaussianblur)
plt.title('gaussian blurred image')
plt.show()
```

```python
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])

image_sharp = cv2.filter2D(image, -1, kernel)

plt.imshow(image_sharp)
plt.title('sharpened image')
plt.show()
```

```python
imageg = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

image_enhanced = cv2.equalizeHist(imageg)

plt.imshow(imageg, 'gray')
plt.title('original image')
plt.show()

plt.imshow(image_enhanced, 'gray')
plt.title('enhanced image')
plt.show()
```

```python
imagey[:, :, 0] = cv2.equalizeHist(imagey[:, :, 0])

imageyr = cv2.cvtColor(imagey, cv2.COLOR_YUV2RGB)

plt.imshow(imagey)
plt.title('original image')
plt.show()

plt.imshow(imageyr)
plt.title('equlizehisted image')
plt.show()
```

```python
max_output = 255
neighborhood_size = 99
subtract_from_mean = 10

image_binarized = cv2.adaptiveThreshold(imageg, 
                                        max_output,
                                        cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                        cv2.THRESH_BINARY,
                                        neighborhood_size,
                                        subtract_from_mean)

plt.imshow(image_binarized, 'gray')
plt.title('adaptive threshold image')
plt.show()
```
