rectangle 그리기

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

img_rectangle = np.ones((400, 400), dtype = 'uint8')

cv2.rectangle(img_rectangle, (50,50), (300,300), (255,255,255), -1)

plt.imshow(img_rectangle, 'gray') # 알아서 linear scaling됨
plt.show()
```

circle 그리기

```python
img_circle = np.ones((400, 400), dtype='uint8')
cv2.circle(img_circle, (300,300), 50, (255,255,255), -1)

plt.imshow(img_circle, 'gray')
plt.show()
```

rectangle과 circle을 bitwise

```python
bitwiseAnd = cv2.bitwise_and(img_rectangle, img_circle)
plt.imshow(bitwiseAnd, 'gray')
plt.title('bitwiseAnd')
plt.show()

bitwiseOr = cv2.bitwise_or(img_rectangle, img_circle)
plt.imshow(bitwiseOr, 'gray')
plt.title('bitwiseOr')
plt.show()

bitwiseXor = cv2.bitwise_xor(img_rectangle, img_circle)
plt.imshow(bitwiseXor, 'gray')
plt.title('bitwiseXor')
plt.show()

bitwiseNot_rec = cv2.bitwise_not(img_rectangle)
plt.imshow(bitwiseNot_rec, 'gray')
plt.title('bitwiseNot_rec')
plt.show()

bitwiseNot_cir = cv2.bitwise_not(img_circle)
plt.imshow(bitwiseNot_cir, 'gray')
plt.title('bitwiseNot_cir')
plt.show()
```

rectangle 여러 개 그리기

```python
img_rectangle = np.zeros((500, 1000), dtype='uint8')

cv2.rectangle(img_rectangle, (60,50), (280,280), (255,255,255), -1)
cv2.rectangle(img_rectangle, (420,50), (550,230), (255,255,255), -1)
cv2.rectangle(img_rectangle, (750,50), (920,280), (255,255,255), -1)

plt.imshow(img_rectangle, 'gray')
plt.show()
```
