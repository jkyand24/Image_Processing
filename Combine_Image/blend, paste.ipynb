```python
import cv2
import matplotlib.pyplot as plt
```

```python
sample_img = cv2.imread('./data/ex_image.png')
sample_logo = cv2.imread('./data/ex_image_logo.png')

sample_img = cv2.cvtColor(sample_img, cv2.COLOR_BGR2RGB)
sample_logo = cv2.cvtColor(sample_logo, cv2.COLOR_BGR2RGB)

plt.imshow(sample_img)
plt.title('sample_img')
plt.show()

plt.imshow(sample_logo)
plt.title('sample_logo')
plt.show()

img_re = cv2.resize(sample_img, (800, 600))
logo_re = cv2.resize(sample_logo, (800, 600))
```

```python
for (alpha, beta) in [(0.5, 0.5), (0.8, 0.2), (1, 1)]:
    blended = cv2.addWeighted(img_re, alpha, logo_re, beta, 0)
    # (imgA, a, imgB, b, c) -> imgA * a + imgB * b + c

    plt.imshow(blended)
    plt.title(f"blended\n(alpha = {alpha}, beta = {beta})")
    plt.show()
```

==========================================================================

```python
logo_small = cv2.resize(logo_re, (300,300))

plt.imshow(logo_small)
plt.show()

x_offset = 40
y_offset = 170

x_end = x_offset + logo_small.shape[1]
y_end = y_offset + logo_small.shape[0]

img_re[y_offset:y_end, x_offset:x_end] = logo_small

plt.imshow(img_re)
plt.show()
```
