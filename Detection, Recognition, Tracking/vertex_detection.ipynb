```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_path = './data/box_image.jpeg'

image = cv2.imread(image_path)

imageg = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
imageg = np.float32(imageg)
```

```python
block_size = 4
aperture = 29
free_parameter = 0.04

detector_responses = cv2.cornerHarris(imageg,
                                      block_size,
                                      aperture,
                                      free_parameter)
detector_responses = cv2.dilate(detector_responses, None)

plt.imshow(detector_responses)
plt.title('detector_responses')
plt.show()

threshold = 0.01
for i in range(detector_responses.shape[0]):
    for j in range(detector_responses.shape[1]):
        if detector_responses[i, j] > threshold * detector_responses.max():
            cv2.circle(image, (j, i), radius = 5, color=(0, 0, 255), thickness=-1)
            
plt.imshow(image)
plt.title('cornerHarris')
plt.show()
            
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            
plt.imshow(image)
plt.title('cornerHarris')
plt.show()
```
