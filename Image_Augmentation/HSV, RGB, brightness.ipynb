```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

ori_image = cv2.imread("./data/image02.jpeg")
ori_image = cv2.cvtColor(ori_image, cv2.COLOR_BGR2RGB)
```

HSV

```python
# RGB -> HSV

hsv_image = cv2.cvtColor(ori_image, cv2.COLOR_RGB2HSV)

# transformation

hsv_image[..., 0] = (hsv_image[..., 0] + 50) % 180
hsv_image[..., 1] = np.uint8(np.float32(hsv_image[..., 1]) * 0.5)
hsv_image[..., 2] = np.uint8(np.float32(hsv_image[..., 2]) * 1.5)

# HSV -> RGB

rgb_image = cv2.cvtColor(hsv_image, cv2.COLOR_HSV2RGB)

plt.imshow(rgb_image)
plt.title('rgb_image')
plt.show()
```

RGB

```python
rgb_image = ori_image.copy()

balance = [0.8, 0.7, 0.8]

for i in range(3):
    if balance[i] != 1.0:
        rgb_image[..., i] = cv2.addWeighted(rgb_image[..., i], balance[i], 0, 0, 0)
        
plt.imshow(rgb_image)
plt.title('rgb_image')
plt.show()
```

Brightness

```python
brightness_offset = 50

bright_image = cv2.convertScaleAbs(ori_image, alpha=1, beta=brightness_offset)

plt.imshow(bright_image)
plt.title('bright_image')
plt.show()
```
