```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

ori_image = cv2.imread("./data/image02.jpeg")
ori_image = cv2.cvtColor(ori_image, cv2.COLOR_BGR2RGB)

grayscale_image = cv2.cvtColor(ori_image, cv2.COLOR_RGB2GRAY)
```

histogram equalization

```python
he_image = cv2.equalizeHist(grayscale_image)

plt.imshow(he_image, 'gray')
plt.title('he_image')
plt.show()
```

adaptive threshold

```python
at_image = cv2.adaptiveThreshold(grayscale_image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)

plt.imshow(at_image, 'gray')
plt.title('at_image')
plt.show()
```
