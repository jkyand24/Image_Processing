```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

ori_image = cv2.imread("./data/image02.jpeg")
ori_image = cv2.cvtColor(ori_image, cv2.COLOR_BGR2RGB)
```

Salt & Pepper noise

```python
sp_noise = np.random.randint(0, 256, ori_image.shape)
black = sp_noise < 20
white = sp_noise > 235

sp_noise_image = ori_image.copy()

sp_noise_image[black] = sp_noise_image[black] * 0
sp_noise_image[white] = sp_noise_image[white] + 255

np.clip(sp_noise_image, 0, 255)

plt.imshow(sp_noise_image)
plt.title('sp_noise_image')
plt.show()
```

Gaussian noise

```python
mean = 0
sigma = 100 ** 0.5

gauss_noise = np.random.normal(mean, sigma, ori_image.shape).astype('uint8')

gauss_noise_image = cv2.add(ori_image, gauss_noise)

plt.imshow(gauss_noise_image)
plt.title('gauss_noise_image')
plt.show()

gauss_noise_image2 = ori_image + gauss_noise

plt.imshow(gauss_noise_image2)
plt.title('gauss_noise_image2')
plt.show()
```
