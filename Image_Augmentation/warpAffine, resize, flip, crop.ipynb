```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

ori_image = cv2.imread("./data/image02.jpeg")
ori_image = cv2.cvtColor(ori_image, cv2.COLOR_BGR2RGB)
```

warpAffine()

```python
#  

shift = (0, 50)

M = np.float32([[1, 0, shift[0]],
                [0, 1, shift[1]]])

shifted_image = cv2.warpAffine(ori_image, M, (ori_image.shape[1], ori_image.shape[0]))

plt.imshow(shifted_image)
plt.title('shifted_image')
plt.show()

#  

M = cv2.getRotationMatrix2D(center=(ori_image.shape[1] / 2, ori_image.shape[0] / 2), 
                            angle=45, 
                            scale=1)

rotated_image = cv2.warpAffine(ori_image, M, (ori_image.shape[1], ori_image.shape[0]))

plt.imshow(rotated_image)
plt.title('rotated_image')
plt.show()
```

resize()

```python
resize_scale = 4

resized_image = cv2.resize(ori_image, 
                           (ori_image.shape[1] * resize_scale, ori_image.shape[0] * resize_scale), 
                           interpolation=cv2.INTER_CUBIC)

plt.imshow(resized_image)
plt.title('resized_image')
plt.show()
```

flip()

```python
flipped_image = cv2.flip(ori_image, 1) # 0: , 1: 

plt.imshow(flipped_image)
plt.title('flipped_image')
plt.show()
```

 crop

```python
cx, cy, cw, ch = 100, 100, 200, 200

cropped_image = ori_image[cy:(cy+ch), cx:(cx+cw)]

plt.imshow(cropped_image)
plt.title('cropped_image')
plt.show()
```
